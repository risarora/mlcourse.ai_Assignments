7. Is it true that Switzerland won fewer medals than Serbia at the 2016 Olympics? Do not consider NaN values in Medal column.

```python
Switzerland = data[(data['Team']=='Switzerland')&(data['Year']==2016)& (data['Medal'].notnull())]['Medal'].count()
Serbia = data[(data['Team']=='Serbia')&(data['Year']==2016)& ( data['Medal'].notnull())]['Medal'].count()
print("Switzerland < Serbia : ",Switzerland < Serbia)
```
